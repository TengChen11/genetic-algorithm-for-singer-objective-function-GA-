```python
from function import *
import copy

variable_size = 2
population_size = 100
x_upper = np.array([5.12, 5.12])
x_lower = np.array([-5.12, -5.12])
crossover_eta = 1
mutation_eta = 1 
crossover_p = 0.4
is_crossover = False
mutation_p = 1 / population_size
is_mutation = False
elite_rate = 0.3
save_path1 = r'./fitness_results_f1'
save_path2 = r'./output_f1'
iteration_num = 100
minimum = 0
minimum_lst = []
maximum_lst = []
mean_lst = []
entire_evolution = []
box_num = 30
iteration1 = 3
iteration2 = 4

population = initialize_population(population_size, variable_size, x_upper, x_lower)
entire_evolution.append(copy.deepcopy(population))
minimum, maximum, mean = fitness_plot(population, save_path1, 0, f1)
minimum_lst.append(minimum)
maximum_lst.append(maximum)
mean_lst.append(mean)
for num in range(1, iteration_num):
    #crossover
    for idx in range(len(population) - 1):
        offspring1, offspring2, is_crossover = SBX(crossover_p, crossover_eta, x_upper, x_lower, population[idx], population[idx + 1], is_crossover)
        if is_crossover:
            population.append(offspring1)
            population.append(offspring2)

    #mutation
    for individual in population:
        offspring, is_mutation = poly_mutation(mutation_p, mutation_eta, x_upper, x_lower, individual, is_mutation)
        if is_mutation:
            population.append(offspring)

    #selection
    population = selection(population_size, population, elite_rate, f1)
    entire_evolution.append(copy.deepcopy(population))

    #plot
    minimum, maximum, mean = fitness_plot(population, save_path1, num, f1)
    minimum_lst.append(minimum)
    maximum_lst.append(maximum)
    mean_lst.append(mean)
minimum_fitness_plot(minimum_lst, save_path2)
max_mean_fitness_plot(maximum_lst, mean_lst, save_path2)
evolution_plot(entire_evolution[iteration1], iteration1, entire_evolution[iteration2], iteration2, save_path2, f1)
boxplot(box_num, entire_evolution, save_path2, f1)
```
